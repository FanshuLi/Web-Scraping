# 1. Coingecko(CKO) Web Scarping

```python
browser = webdriver.Chrome(executable_path="/Users/kristy/Desktop/BitMart_new/10.爬虫/10.29 button爬虫/chromedriver")
browser.get("https://www.coingecko.com/en/exchanges")
```

```python
soup = BeautifulSoup(browser.page_source,'html.parser')
```

```python
def cko():
    rank=[]
    exchange=[]
    trust_score=[]
    volume_normalized=[]
    volume=[]
    website_visits=[]
    coins=[]
    pairs=[]


    for tr in soup.find_all('tr'):
        try:

            tds = tr.find_all('td')
    #         
            if not tds:
                pass
            else:

                rank.append(tds[0].text.strip('\n').strip('\n '))
                exchange.append(tds[1].text.strip('\n').strip('\n').split('\n')[0])
                trust_score.append(tds[2].text.strip('\n').strip('\n '))
                volume_normalized.append(tds[3].text.strip('\n').strip('\n ').strip('$').replace(',',''))
                volume.append(tds[4].text.strip('\n').strip('\n ').strip('\n\n\n\n Trading Incentives').strip('$').replace(',',''))
                website_visits.append(tds[5].text.strip('\n').strip('\n ').replace(',',''))
                coins.append(tds[6].text.strip('\n').strip('\n '))
                pairs.append(tds[7].text.strip('\n').strip('\n '))
        except:
            print('!')
            pass
    
    return rank,exchange,volume_normalized,volume,website_visits
    
```

```python
b=cko()
```

```python
# save to dataframe
cko=pd.DataFrame({'cko_rank':b[0],'exchange':b[1],'cko_24hVolume':b[3],
                  'website_visits':b[4]})
```

```python
cko['daily_visits']=cko['website_visits'].astype(int)/30
cko['exchange']=cko['exchange'].str.lower()
```

# 2. CoinMarketCap(CMC) Web Scarping

```python
# Get page
page_url='https://coinmarketcap.com/rankings/exchanges/reported/'
page = requests.get(page_url) 
soup = BeautifulSoup(page.content,'html.parser')
```

```python
# Define a function 

def cmc(n):
    
    
    Rank=[]
    Exchange=[]
    Volume_24h=[]
    Volume_7d=[]
    Volume_30d=[]

    
    
    for i in n:
        j=str(i)
        page_url = "https://coinmarketcap.com/rankings/exchanges/reported/"+j+""
        print(page_url)
        
        page = requests.get(page_url) 

        if page.status_code==200:
            
            soup = BeautifulSoup(page.content, 'html.parser')
    
            
            rank=[]
            exchange=[]
            volume_24h=[]
            volume_7d=[]
            volume_30d=[]



            for tr in soup.find_all('tr'):
                try:

                    tds = tr.find_all('td')
            #         
                    if not tds:
                        pass
                    else:
                        #print(tds[3].text)
                        exchange.append(tds[1].text.strip('\n').strip('\n '))
                        volume_24h.append(tds[2].text.strip('\n').strip('\n ').strip('$').replace(',',''))
                        volume_7d.append(tds[3].text.strip('\n').strip('\n ').strip('$').replace(',',''))
                        volume_30d.append(tds[4].text.strip('\n').strip('\n ').strip('$').replace(',',''))
                        rank.append(tds[0].text.strip('\n').strip('\n '))


                except:
                    print('!')
                    pass
                
        Rank.extend(rank)
        Exchange.extend(exchange)
        Volume_24h.extend(volume_24h)
        Volume_7d.extend(volume_7d)
        Volume_30d.extend(volume_30d)
        

        
        
    return Rank,Exchange,Volume_24h,Volume_7d,Volume_30d
```

```python
a=cmc(range(1,3))
```

```python
# save to dataframe
cmc=pd.DataFrame({'cmc_rank':a[0],'exchange':a[1],'cmc_30dVolume':a[4]})
```

# 3. Data Integration 

```python
inner=pd.merge(cko,cmc,on='exchange',how='inner')
inner['cko_24hVolume']=inner['cko_24hVolume'].astype(float)
```

```python
inner.count()
```

```python
inner.head()
```

```python
# Average
inner['avg']=(inner['cko_24hVolume'] + inner['cmc_avgVolume'])/2
```

```python
# Calculate ADUTV
inner['ADUTV']=inner['avg']/inner['daily_visits']
```

```python
inner.loc[inner['exchange'].isin(['binance', 'bitfinex', 'kraken', 
                                'bitstamp', 'coinbase pro', 'bitflyer',
                                'gemini', 'itbit', 'bittrex', 
                                'poloniex'])].reset_index(drop=True)
```

```python
# Calculate Benchmark
bench=np.median(inner['ADUTV'].loc[inner['exchange'].isin(['binance', 'bitfinex', 'kraken', 
                                                           'bitstamp', 'coinbase pro', 'bitflyer',
                                                           'gemini', 'itbit', 'bittrex', 
                                                           'poloniex'])])
```

```python
bench
```

```python
adjust_volume=[]
for i,j in zip(inner['ADUTV'],inner['daily_visits']):
    if i<= bench:
        adjust_volume.append(i)
    else:
        adjust_volume.append (bench*j)
```

```python
inner['adjust_volume']=adjust_volume
```

```python
inner=inner[['cko_rank','cmc_rank','exchange','website_visits','adjust_volume']]
```
